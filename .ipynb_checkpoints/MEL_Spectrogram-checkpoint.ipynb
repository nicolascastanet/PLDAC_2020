```python
import pandas as pd
from PIL import Image
import librosa
import numpy as np
import librosa.display
import matplotlib.pyplot as plt 

DATA_PATH = 'ff1010bird_metadata.csv'
```

- chargement des labels

```python
df = pd.read_csv(DATA_PATH)
print(df.shape)

i=0
for row in df.iterrows():
    print("----------------------------")
    print("recording id ->",row[1][0])
    print("recording label ->",row[1][1])
    i+=1
    if i%5 == 0:
        break
        
df.head()
```

## Visualisation du spectrogramme d'un chant d'oiseau

```python
bird_sound = df.where(df['hasbird'] == 1)
bird_sound.dropna()
```

```python
#exemple de l'enregistrement 19037 labellé +1
filename = 'wav/97375.wav'
```

- représentation temporelle

```python
y, sr = librosa.load(filename)
# trim silent edges
bird_song, _ = librosa.effects.trim(y)
librosa.display.waveplot(bird_song, sr=sr)
```

- représentation fréquentielle

```python
n_fft = 2048
D = np.abs(librosa.stft(bird_song[:n_fft], n_fft=n_fft,
                        hop_length=n_fft+1))
plt.plot(D)
```

- Spectrogram

```python
hop_length = 1024
D = np.abs(librosa.stft(bird_song, n_fft=n_fft,  
                        hop_length=hop_length))
DB = librosa.amplitude_to_db(D, ref=np.max)
librosa.display.specshow(DB, sr=sr, x_axis='time', y_axis='log');
plt.colorbar(format='%+2.0f dB')
print(DB.shape)
```

```python
mel_bins = 128
mel = librosa.filters.mel(sr=sr, n_fft=n_fft, n_mels=mel_bins)
```

```python
plt.figure(figsize=(15, 4))
plt.subplot(1, 3, 1)
librosa.display.specshow(mel, sr=sr, hop_length=hop_length, 
                         x_axis='linear')
plt.ylabel('Mel filter')
plt.colorbar()

plt.title('1. Our filter bank for converting from Hz to mels.')
plt.subplot(1, 3, 2)
mel_10 = librosa.filters.mel(sr=sr, n_fft=n_fft, n_mels=10)
librosa.display.specshow(mel_10, sr=sr, hop_length=hop_length, 
                         x_axis='linear')
plt.ylabel('Mel filter')
plt.colorbar()
plt.title('2. Easier to see what is happening with only 10 mels.')
plt.subplot(1, 3, 3)
idxs_to_plot = [0, 9, 49, 99, 127]
for i in idxs_to_plot:
    plt.plot(mel[i])
plt.legend(labels=['{}'.format(i+1) for i in idxs_to_plot])
plt.title('3. Plotting some triangular filters separately.')
plt.tight_layout()
```

```python
S = librosa.feature.melspectrogram(bird_song, sr=sr, n_fft=n_fft, 
                                   hop_length=hop_length, 
                                   n_mels=mel_bins)
S_DB = librosa.power_to_db(S, ref=np.max)
librosa.display.specshow(S_DB, sr=sr, hop_length=hop_length, 
                         x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
print(S_DB.shape)
```
